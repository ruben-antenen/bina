```python
# prompt: bitte lade alee daten aus https://raw.githubusercontent.com/alendoko/bina/refs/heads/main/LB1/DATA/NYC_Weather_2016_2022.csv

import pandas as pd

# Load the data from the URL
url = "https://raw.githubusercontent.com/alendoko/bina/refs/heads/main/LB1/DATA/NYC_Weather_2016_2022.csv"
try:
  df = pd.read_csv(url)
  print("Data loaded successfully.")
  # You can now work with the DataFrame 'df'
except Exception as e:
  print(f"An error occurred while loading the data: {e}")
```

```python
df.head()
```

```python
# prompt: bitte erstelle einen neuen dataframe df_noon (aufgrund vom df) wo du nur reihen übernimmst wenn es ein wert für 12:00 Uhr mittag ist, The column with the time is called time, example of time 2016-01-01T02:00

# Convert the 'time' column to datetime objects
df['time'] = pd.to_datetime(df['time'])

# Extract the hour from the 'time' column
df['hour'] = df['time'].dt.hour

# Create a new DataFrame containing only rows where the hour is 12
df_noon = df[df['hour'] == 12]

# You can now work with the DataFrame 'df_noon' which contains only data from 12:00
print(df_noon.info())
df_noon.tail()
```

```python
# prompt: nimm df_noon und gib mir erstelle ein neues dataframe df_noon_2016_2017 wo du nur die zwei jahre 2016 und 2017 nimmst

import pandas as pd

# Load the data from the URL (assuming the previous code was executed)
# If not, execute the previous code block first.
try:
    df_noon_2016_2017 = df_noon[(df_noon['time'].dt.year == 2016) | (df_noon['time'].dt.year == 2017)]
    print("DataFrame 'df_noon_2016_2017' created successfully.")
    print(df_noon_2016_2017.info())
    print(df_noon_2016_2017.head())
except NameError:
    print("Error: 'df_noon' not found. Please run the previous code block first.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df_noon_2016_2017.to_csv('NYC_Weather_2016_2017_noon.csv', index=False)
```
